```python
import configparser
config = configparser.ConfigParser()
config.read('C:/Users/etjones/Desktop/AI OPS/test1/model.cfg')
config.sections()
nodes=config.get('nodes', 'names')
supply=config.get('define', 'SUPPLY')
demand=config.get('define','DEMAND')

def convlist(string):
    li = list(string.split(", "))
    return li

nodes = convlist(nodes)
supply = convlist(supply)
demand = convlist(demand)

def convdict(lst):
    res_dct = {i: lst[i] for i in range(0, len(lst))}
    return res_dct

dict_nodes = convdict(nodes)
dict_supply = convdict(supply)
dict_demand = convdict(demand)

start_nodes = []
end_nodes = []

for ii in range(len(supply)):
    start_nodes.append(0)

for ii in range(len(dict_nodes)):
    if dict_nodes[ii] in supply:       
        for jj in range(len(demand)):
            start_nodes.append(ii)
    if dict_nodes[ii] in demand:
        start_nodes.append(ii)
            
for ii in range(1,len(supply)+1):
    end_nodes.append(ii)

for ii in range(len(dict_nodes)):
    if dict_nodes[ii] in demand:
        for jj in range(len(supply)):
            end_nodes.append(ii)

    if dict_nodes[ii] == 'Sink':
        for jj in range(len(demand)):
            end_nodes.append(ii)

print("Start Nodes:", start_nodes)
print("End Nodes:", end_nodes)     
```
